On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

1. Think about what could be going wrong with our calculation. Think about a better way to evaluate this data.
2. What metric would you report for this dataset?
3. What is its value?

Something that should have been taken into consideration when calculating the average order value are the outliers since they will skew the value. 

An alternative metric to report for this dataset is to use the median value since it handles the outliers better and is often a better representation of the average order value. To calculate the median value, the data must first be sorted by the order value to be able to find the middle. 

The median order amount is $284.00 which is a better representation of the AOV for sneaker purchases. 


```python
import pandas as pd

def get_median_order_value (filename):
    median = 0
    column = 'order_amount'
    df = pd.read_csv(filename)
    median = df[column].median() 
    return median

aov = get_median_order_value('./dataset.csv')

print (aov)
```

**How many orders were shipped by Speedy Express in total?**

**What is the last name of the employee with the most orders?**

Answer: Peacock 

SELECT LastName, COUNT() as TotalOrders FROM Orders 
INNER JOIN Employees ON Employees.EmployeeID = Orders.EmployeeID
GROUP BY Employees.EmployeeID 
ORDER BY TotalOrders DESC LIMIT 1;

**What product was ordered the most by customers in Germany?**

Answer: Boston Crab Meat

SELECT Products.ProductName, SUM(Quantity) AS TotalOrders 
FROM Orders 

JOIN OrderDetails ON OrderDetails.OrderID = Orders.OrderID 

JOIN Products ON Products.ProductID = OrderDetails.ProductID 

JOIN Customers ON Customers.CustomerID = Orders.CustomerID 
WHERE Customers.Country = "Germany"
GROUP BY OrderDetails.ProductId 
ORDER BY TotalOrders DESC LIMIT 1
